Outstate and S.F. Ration seem to be good predictor of target label. Their discributions are still overlapping but it looks that they split the data enough..

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(df.drop('Private', axis=1))
```

```python
kmeans.cluster_centers_
```

## Evaluation

Due to present of labels in dataset, we are able tu evaluate the performance of model.

```python
df['Cluster'] = df['Private'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df['Cluster'],kmeans.labels_))
print(classification_report(df['Cluster'],kmeans.labels_))
```
